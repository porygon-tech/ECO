```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x, y = np.meshgrid(np.arange(0, 3, 0.1), np.arange(0, 3, 0.1))
dx =   x - x * y
dy = - y + x * y

fig = plt.figure(figsize = (12,9))
plt.streamplot(x, y, dx, dy, density=2.5, linewidth=0.8, arrowsize=0.8)
plt.show()
```

```python
delta=0.01
duration=10

def xdot (x,y):
  return x-x*y

def ydot (x,y):
  return -y + x*y
```

```python
x,y = np.random.rand(2)*3
ntimesteps=int(duration/delta)
timeseries=np.zeros((2,ntimesteps))
for t in range(ntimesteps):
  x+=xdot(x,y)*delta
  y+=ydot(x,y)*delta
  timeseries[0,t] = x
  timeseries[1,t] = y
```

```python
fig = plt.figure(); ax = fig.add_subplot(111)
ax.plot(timeseries.T)
plt.show()
```

```python
fig = plt.figure(); ax = fig.add_subplot(111)
ax.plot(timeseries[0,:],timeseries[1,:])
plt.show()
```

```python
nsamples=20
ntimesteps=int(duration/delta)
timeseries=np.zeros((2,ntimesteps,nsamples))
for s in range(nsamples):
  x,y = np.random.rand(2)*3
  for t in range(ntimesteps):
    x+=xdot(x,y)*delta
    y+=ydot(x,y)*delta
    timeseries[0,t,s] = x
    timeseries[1,t,s] = y
```

```python
fig = plt.figure(); ax = fig.add_subplot(111)
ax.plot(timeseries[0,:],timeseries[1,:])
plt.show()
```
