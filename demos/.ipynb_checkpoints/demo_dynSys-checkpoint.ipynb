```python
from pylab import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
def samplecolors(n, type='hex',palette=plt.cm.gnuplot):
    if type == 'hex':
        return list(map(plt.cm.colors.rgb2hex, list(map(palette, np.linspace(1,0,n)))))
    elif type == 'rgba':
        return list(map(palette, np.linspace(1,0,n)))
```

```python



x, y = meshgrid(arange(0, 3, 0.1), arange(0, 3, 0.1))
xdot = x - x * y
ydot = - y + x * y

streamplot(x, y, xdot, ydot, density=1.5, linewidth=0.8, arrowsize=0.5)
show()
```

```python
delta=0.01
duration=10

def xdot (x,y):
  return x-x*y

def ydot (x,y):
  return -y + x*y
```

```python
x,y = np.random.rand(2)*3
ntimesteps=int(duration/delta)
timeseries=np.zeros((2,ntimesteps))
for t in range(ntimesteps):
  x+=xdot(x,y)*delta
  y+=ydot(x,y)*delta
  timeseries[0,t] = x
  timeseries[1,t] = y
```

```python
fig = plt.figure(); ax = fig.add_subplot(111)
ax.plot(timeseries.T)
plt.show()
```

```python
fig = plt.figure(); ax = fig.add_subplot(111)
ax.plot(timeseries[0,:],timeseries[1,:])
plt.show()
```

```python
nsamples=20
ntimesteps=int(duration/delta)
timeseries=np.zeros((2,ntimesteps,nsamples))
for s in range(nsamples):
  x,y = np.random.rand(2)*3
  for t in range(ntimesteps):
    x+=xdot(x,y)*delta
    y+=ydot(x,y)*delta
    timeseries[0,t,s] = x
    timeseries[1,t,s] = y
```
